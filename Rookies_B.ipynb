## **Question**: Factors that makes the difference (in literacy rate) between Kerala and Bihar?



```python
# Extracting KERALA and BIHAR litercacy data from state elementary data 
st_el_kb=st_el.iloc[[9,31],[2,7,8,10,11,14,15,16,37,46,28,55,64,205,245,229,253,261,293]]

# calulating female percentage population for KERALA and BIHAR
st_el_kb["FEMALEPOP%"]=(st_el["SEXRATIO"]/(st_el["SEXRATIO"]+1000))*100

# calulating total urban population from state elementary data in KERALA and BIHAR
st_el_kb["TOT_URB_POP"]=(st_el["TOTPOPULAT"]*st_el["P_URB_POP"])/100

# calculating total rural population in KERALA and BIHAR
st_el_kb["TOTRPOP"]=st_el_kb["TOTPOPULAT"]-st_el_kb["TOT_URB_POP"]

# printing data set
st_el_kb
```

# FROM BELOW GRAPH
    # i) Total population count in BIHAR much higher than KERALA
    **Solutions**: Population controle measure.
    # ii)BIHAR has much higher rural population percantage than KERALA 
    # iii) BIHAR has less URBAN Population percentage as compared to KERALA

```python

# plot state wise population distribution in rual and urban area
st_el_kb.plot(x="STATNAME",y=["TOTRPOP","TOTPOPULAT","TOT_URB_POP"],kind="bar",fontsize=10,ylabel="Population size",title="Population distribution across rural and urban area of KERALA and BIHAR")
#Display graph
plt.show()
```

# FROM BELOW GRAPH
    # KERALA has much higher literacy rate and better FEMALE Population percentage as compared to BIHAR
    **Suggestion**: measure to controle controle female foeticide and increase female enrollment in school

```python
#Plot state and gender wise literacy from state elementary data 
#STATNAME: state name
#OVERALL_LI: overall literacy rate 
#FEMALE_LIT: female literacy rate 
#MALE_LIT: Male literacy rate
#FEMALEPOP: female population percentage

st_el_kb.plot(x="STATNAME",y=["OVERALL_LI", "FEMALE_LIT","MALE_LIT","FEMALEPOP%"],kind="bar",fontsize=10,ylabel="Percentage(%)",title="Percentage literacy rate and percentage female population in KERALA and BIHAR")
# Draw graph
plt.show()
```

# **Question:** Area wise total primary school and area wise total primary enrollement 

```python
# Extracting statewise litercacy data from state elementary data

st_el1=st_el.iloc[[9,31],[2,7,8,109,365]]
#st_el1
# Extracting statewise litercacy data from state secondary data
st_se1=st_se.iloc[[9,31],[18,25,32,46,342]]
#st_se1
# Mergin secondary data to elementary table
st_el1["TOT_PRI_SCH"]=st_se1["sch_1"]
st_el1["TOT_PRI_RU_SCH"]=st_se1["sch_r_1"]
st_el1["TOT_PRI_UR_SCH"]=st_se1["sch_u_1"]
st_el1["TOT_RU_ENR"]=st_se1["enr_all"]
st_el1["TOT_GRL_EN"]=st_se1["enrg_all"]
#st_el1
#  Calulating total boys enrollment and boys rural enrollment
st_el1["ENRBTOT"]=st_el1["ENRTOT"]-st_el1["ENRGTOT"]
st_el1["TOT_RU_B_EN"]=st_el1["TOT_RU_ENR"]-st_el1["TOT_GRL_EN"]
st_el1["TOT_UR_EN"]=st_el1["ENRTOT"]-st_el1["TOT_RU_ENR"]

#display data 
st_el1
```

# **FROM BELOW GRAPH**
 i) Total rural enrollement in bihar is very less while in KERALA it is proportional
    **solutions:** Increase rural enrollment in BIHAR

```python
# Plotting graph for enrollment across rural and urban area in primary school in KERALA and BIHAR
st_el1.plot(x="STATNAME",y=["TOT_RU_ENR","ENRTOT","TOT_UR_EN"],kind="bar",ylabel="Total enrollement in Lakh",title="Total rural and urban enrollement in BIHAR and KERALA")
```
